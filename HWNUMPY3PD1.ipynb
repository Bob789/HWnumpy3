```python
import numpy as np
#solution 1 a
np.random.seed(80)
rand_array = np.random.randint(1, 101, (4, 4))
print(rand_array)
ra = rand_array.copy()
ra = ra[(ra > 50)]
print(ra)
```

```python
#solution 1 b
print('SUM of columns ra array axis=0 columns:\n', rand_array.sum(axis=0))
print('SUM of rows ra array axis=0 columns:\n', rand_array.sum(axis=1))
```

```python
#solution 1 c
print('MAX of COLUMNS ra array  : ',np.max(rand_array,axis=0))
print('MAX of ROWS ra array     : ',np.max(rand_array,axis=1))
print('MIN of COLUMNS ra array  : ',np.min(rand_array,axis=0))
print('MIN of ROWS ra array     : ',np.min(rand_array,axis=1))
print('MEAN of COLUMNS ra array : ',np.mean(rand_array,axis=0))
print('MEAN of ROWS ra array    : ',np.mean(rand_array,axis=1))
```

```python
#solution 2
import pandas as pd
from IPython.display import display

new_games = {
    "Final Fantasy VII Rebirth": 69.99,
    "Astro's Playroom"         : 69.99,
    "Astro Bot"                : 69.99,
    "Metaphor: ReFantazio"     : 59.99,
    "Final Fantasy XVI"        : 59.99,
    "Starfield"                : 69.99,
    "Hades II"                 : 49.99
}

games_series = pd.Series(new_games)

print("Metaphor: ReFantazio price: ", games_series["Metaphor: ReFantazio"])
print("The indices are:")
for idx in games_series.index:
    print("-", idx)
print("Value of index 3 are : ", games_series.iloc[3])
print("Sum of the values :", games_series.sum())
```
